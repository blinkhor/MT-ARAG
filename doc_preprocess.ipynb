```python
import pandas as pd
import numpy as np
import json
```

```python
music_therapy_data = pd.read_excel('docs/preset_documents.xlsx')
```

```python
# 自定义转换函数
def convert_value(value):
    if pd.isna(value):  # 处理NaN/空值
        return None
    elif isinstance(value, (np.bool_, bool)):  # 处理布尔值
        return bool(value)
    elif isinstance(value, (np.int64, np.float64)):  # 处理numpy数值类型
        return int(value) if isinstance(value, np.int64) else float(value)
    return value

# 遍历DataFrame的每一行
for index, row in music_therapy_data.iterrows():
    # 将行数据转换为字典并处理特殊值
    row_dict = {key: convert_value(value) for key, value in row.to_dict().items()}
    
    # 定义JSON文件名
    json_filename = f'docs/music_therapy_row_{index}.json'
    
    # 将字典写入JSON文件
    with open(json_filename, 'w', encoding='utf-8') as f:
        json.dump(row_dict, f, ensure_ascii=False, indent=4)
```
